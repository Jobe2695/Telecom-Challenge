#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
datos = pd.read_json(url)
datos
```

```python
#Columnas con diccionarios
dict_columns = ['customer', 'phone', 'internet', 'account']

# Creae una copia de trabajo
df_expanded = datos.copy()

for col in dict_columns:
    #Normalizar la columnas
    normalized_col_df = pd.json_normalize(datos[col])
    # Agregar un prefijo al original
    normalized_col_df = normalized_col_df.add_prefix(f'{col}_')
    # Concatenar con el dataFrame
    df_expanded = pd.concat([df_expanded, normalized_col_df], axis=1)

# Borrar columanas originales que fueron expandidas
df_expanded = df_expanded.drop(columns=dict_columns)

# ActualizaciÃ³n de los datos
datos = df_expanded

datos
```
